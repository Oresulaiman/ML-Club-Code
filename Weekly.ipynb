```
#import libraries 
import numpy as np
import pandas as pd

#Load data
cell_phones = pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv')
population = pd.read_csv('https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv')
geo_country_codes = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
cell_phones
```

```
population
```

```
geo_country_codes
```

```
#join cellphones and population dataframe 
cell_phones_pop = cell_phones.merge(population, on=['Country', 'year'], how='inner')
cell_phones_pop
```

```
#rename columns in geo_country_codes, country = geo; name = country
geo_country = geo_country_codes.rename(columns={'country': 'geo', 'name': 'Country'})
geo_country

```

```
#merge ‘geo’ and ‘Country’ columns from the ‘geo_country’ and your ‘merged population and cell phone’ data.
new_geo = geo_country[['geo', 'Country']]
new_geo
new_df = cell_phones_pop.merge(new_geo, on=['Country'])
new_df

```

```
#Calculate the number of cell phones per person, and add this column onto your dataframe.
new_df.dtypes
new_df['cell_per_person'] = new_df.cell_phones_total/new_df.population
new_df[new_df.Country == 'United States'].tail()

```

```
#Modify the ‘geo’ column to make the ‘geocodes' uppercase instead of lowercase.
new_df.geo = new_df.geo.str.upper()
new_df.head()
```

```
#In 2017, what were the top 5 countries with the most cell phones total?
year = new_df[new_df['year'] == 2017]
first = year[['Country', 'cell_phones_total']].groupby('Country').first()
sort = first.sort_values('cell_phones_total', ascending = False)
sort.head(5)
```
